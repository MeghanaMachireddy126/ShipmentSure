```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  # Also needed for plt.show()

# Load dataset
df = pd.read_csv("../data/Shipping_data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# Countplot for 'delivery_status'
sns.countplot(x="delivery_status", data=df)
plt.show()

# Heatmap of correlations
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```
